```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
 

    
path_dir1 = 'C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\dataset\\helmet_no\\'
path_dir2 = 'C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\dataset\\helmet_yes\\'
 
file_list1 = os.listdir(path_dir1)
file_list2 = os.listdir(path_dir2)
 
file_list1_num = len(file_list1)
file_list2_num = len(file_list2)
 
file_num = file_list1_num + file_list2_num
 
#Pemrosesan gambar awal 
num = 0;
all_img = np.float32(np.zeros((file_num, 224, 224, 3))) 
all_label = np.float64(np.zeros((file_num, 1)))
 
for img_name in file_list1:
    img_path = path_dir1+img_name
    img = load_img(img_path, target_size=(224, 224))
    
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    all_img[num, :, :, :] = x
    
    all_label[num] = 0 # tidak pakai helm
    num = num + 1

for img_name in file_list2:
    img_path = path_dir2+img_name
    img = load_img(img_path, target_size=(224, 224))
    
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    all_img[num, :, :, :] = x
    
    all_label[num] = 1 # pakai helm
    num = num + 1
 
 
# mengacak kumpulan data (untuk memastikan pelatihan yang tepat)
n_elem = all_label.shape[0]
indices = np.random.choice(n_elem, size=n_elem, replace=False)
 
all_label = all_label[indices]
all_img = all_img[indices]
 
# memisahkan set pelatihan dan data test
num_train = int(np.round(all_label.shape[0]*0.8))
num_test = int(np.round(all_label.shape[0]*0.2))
 
train_img = all_img[0:num_train, :, :, :]
test_img = all_img[num_train:, :, :, :] 
 
train_label = all_label[0:num_train]
test_label = all_label[num_train:]
 
 
 
# membuat model dasar yang telah dilatih
IMG_SHAPE = (224, 224, 3)
 
base_model = ResNet50(input_shape=IMG_SHAPE, weights='imagenet', include_top=False)
base_model.trainable = False
base_model.summary()
print("Number of layers in the base model: ", len(base_model.layers))
 
flatten_layer = Flatten()
dense_layer1 = Dense(128, activation='relu')
bn_layer1 = BatchNormalization()
dense_layer2 = Dense(1, activation=tf.nn.sigmoid)
 
model = Sequential([
        base_model,
        flatten_layer,
        dense_layer1,
        bn_layer1,
        dense_layer2,
        ])

base_learning_rate = 0.001
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
    loss='binary_crossentropy',
    metrics=['accuracy'])
model.summary()
 
history = model.fit(train_img, train_label, epochs=10, batch_size=16, validation_data = (test_img, test_label))
 
# save model
model.save("model.h5")
 
print("Saved model to disk") 
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from IPython.display import display, Image
import matplotlib.pyplot as plt
import os
import random

# Loadidng model
model = load_model('C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\model.h5')

helmet_yes_path = 'C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\dataset\\helmet_yes\\'
helmet_no_path = 'C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\dataset\\helmet_no\\'

helmet_yes_list = os.listdir(helmet_yes_path)
helmet_no_list = os.listdir(helmet_no_path)

# Randomly select an image from each category
helmet_yes_image = cv2.imread(os.path.join(helmet_yes_path, random.choice(helmet_yes_list)))
helmet_no_image = cv2.imread(os.path.join(helmet_no_path, random.choice(helmet_no_list)))

# Loop through selected images
for frame in [helmet_yes_image, helmet_no_image]:
    # Resize the face region to match the input size expected by the model
    face_region = cv2.resize(frame, (224, 224), interpolation=cv2.INTER_AREA)

    # Preprocess the input image
    x = img_to_array(face_region)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Make a prediction using the model
    prediction = model.predict(x)

    # Determine whether the person is wearing a helmet or not
    if prediction < 0.04:  # Helmet not detected
        label = "Helmet X ({:.2f}%)".format((1 - prediction[0][0]) * 100)
        color = (0, 0, 255)
    else:  # Helmet detected
        label = "Helmet O ({:.2f}%)".format(prediction[0][0] * 100)
        color = (0, 255, 0)

    # Add label to the image
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

    # Display output in Jupyter Notebook
    _, img_encoded = cv2.imencode('.png', frame)
    img_bytes = img_encoded.tobytes()
    display(Image(data=img_bytes))

    # Display Helmet X or Helmet O in separate plots
    plt.figure(figsize=(6, 6))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")

    # Cropped face region
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(face_region, cv2.COLOR_BGR2RGB))
    plt.title(label)
    plt.axis("off")

    plt.show()
```

```python
import cvlib as cv
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import ImageFont, ImageDraw, Image

model = load_model('C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\model.h5')
model.summary()

# open webcam
webcam = cv2.VideoCapture(0)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()

# loop frame
while webcam.isOpened():
    # membaca frame dari webcam
    status, frame = webcam.read()

    if not status:
        print("Could not read frame")
        exit()

    # deteksi wajah
    face, confidence = cv.detect_face(frame)

    # loop untuk wajah yang dideteksi
    for idx, f in enumerate(face):
        (startX, startY) = f[0], f[1]
        (endX, endY) = f[2], f[3]

        if 0 <= startX <= frame.shape[1] and 0 <= endX <= frame.shape[1] and 0 <= startY <= frame.shape[0] and 0 <= endY <= frame.shape[0]:
            face_region = frame[startY:endY, startX:endX]
            face_region1 = cv2.resize(face_region, (224, 224), interpolation=cv2.INTER_AREA)

            x = img_to_array(face_region1)
            x = np.expand_dims(x, axis=0)
            x = preprocess_input(x)

            prediction = model.predict(x)
            threshold = 0.007
            #0.02084012
            #0.01972104
            if prediction < threshold:  # pakai helm
                cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 0, 255), 2)
                Y = startY - 10 if startY - 10 > 10 else startY + 10
                text = "Helmet X ({:.2f}%)".format((1 - prediction[0][0]) * 100)
                cv2.putText(frame, text, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

            else:  # tidak pakai helm
                cv2.rectangle(frame, (startX, startY), (endX, endY), (0, 255, 0), 2)
                Y = startY - 10 if startY - 10 > 10 else startY + 10
                text = "Helmet 0 ({:.2f}%)".format(prediction[0][0] * 100)
                cv2.putText(frame, text, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # output
    cv2.imshow("Helmet classify", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

webcam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from cvlib import detect_face

model = load_model('C:\\Users\\relax\\miniconda3\\envs\\LatihanCV\\helm\\helm\\model.h5')
model.summary()

video_path = input("Enter the video file path: ")

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print("Error: Could not read frame.")
        break

    faces, confidences = detect_face(frame)

    for idx, (startX, startY, endX, endY) in enumerate(faces):
        face_region = frame[startY:endY, startX:endX]

        face_region_resized = cv2.resize(face_region, (224, 224), interpolation=cv2.INTER_AREA)
        x = img_to_array(face_region_resized)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        predictions = model.predict(x)

        # Print predicted values and confidence scores
        print("Predictions:", predictions)
        print("Confidence Scores:", confidences[idx])

        threshold = 0.1  # Adjust threshold as needed
        helmet_detected = predictions[0] > threshold
        if helmet_detected:
            helmet_label = "Helmet O"
            helmet_color = (0, 255, 0)
        else:
            helmet_label = "Helmet X"
            helmet_color = (0, 0, 255)


        cv2.rectangle(frame, (startX, startY), (endX, endY), helmet_color, 2)
        Y = startY - 10 if startY - 10 > 10 else startY + 10
        cv2.putText(frame, helmet_label, (startX, Y), cv2.FONT_HERSHEY_SIMPLEX, 0.7, helmet_color, 2)

    cv2.imshow("Face and Helmet Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
